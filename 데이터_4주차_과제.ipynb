```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False

data = pd.DataFrame(pd.read_csv('기온.csv', encoding='cp949'))

x = pd.to_datetime(data["일시"]).dt.dayofyear

y1 = data["최고기온(°C)"]
y2 = data["평균기온(°C)"]
y3 = data["최저기온(°C)"]


plt.plot(x, y1, color = 'red', linestyle = 'solid', label = '최고기온')
plt.plot(x, y2, color = 'gold', linestyle = 'solid', label = '평균기온')
plt.plot(x, y3, color = 'blue', linestyle = 'solid', label = '최저기온')


plt.legend(loc = 'best', ncol = 1)
plt.xlabel('일')
plt.ylabel('기온')
plt.title('서울시 2022년도 기온 변화')
plt.show()

```
